```python
def wc( text ):
	LoW = text.split()
	print("There are", len(LoW), "words.")
	d = {}
	for w in LoW:
	  if w not in d:
	    d[w] = 1
	  else:
	    d[w] += 1

	print("There are", len(d), "distinct words.\n")
	return d   # this way we can use d later!
```

## Tokenizer

Let's try a different approach with a package. The power of Python is not working alone but leveraging existing libraries.

This time, we try with NTLK tokenizer

### With NTLK

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords') #need to download this pre-trained data for using with word tokenizer
nltk.download('punkt') #need to download this pre-trained data for using with word tokenizer
```

```python
from nltk.tokenize import word_tokenize
stop_words_nltk = set(stopwords.words('english')) #define English as the stop words
```

```python
print(stop_words_nltk)#let's see the stop words
len(stop_words_nltk) #how many stop words are there in this list?
```

```python
tokenized_text_nltk = word_tokenize(text) #tokenize the text with ntlk
tokenized_text_nltk
```

```python
#list have duplicates, you need to remove them.
tokenized_text_nltk = list(dict.fromkeys(tokenized_text_nltk)) #get all the texts to dict, which has no duplicate on the keys, then create a list
len(tokenized_text_nltk) #total words
```

```python
tokenized_text_nltk_wo_sw = [i for i in tokenized_text_nltk if not i in stop_words_nltk] #use list comprehension to remove stop words
tokenized_text_nltk_wo_sw
```

```python
len(tokenized_text_nltk_wo_sw)
```

## Stemming and Lemmatization

Two of the ways to further reduce the number of words but still keep the roots of words are via stemming and lemmatization. Stemming is to cutting off the word while lemmatization is to finding the root word

Example for stemming:

* adjustable --> adjust

* formality --> formaliti (a stemmed token does not need to be a real word)



Example for lemmatization:

* am --> be

* better --> good (a lemma token is a real word)

To create stemming and lemmatization of the word, we lean on NLTK package for this exercise. There are many NLP packages, NLTK is one of them.

### Stemming

NLTK has stemmers.

TextBlob has stemmers too.

Spacy has none.


```python
from nltk.stem.porter import * #import Stemming from NLTK

# Initialize the stemmer
porter = PorterStemmer() #porter is a popular one, there are many more https://www.nltk.org/api/nltk.stem.html

list_of_words = ['run','runner','running','ran','runs','easily','fairly', "cook","cooker","cooking","cooked"]
stem_example = [porter.stem(word) for word in list_of_words]
stem_example
```

```python
stem_nltk = [porter.stem(x) for x in tokenized_text_nltk_wo_sw] #use list comprehension to stem
stem_nltk
```

```python
stem_nltk = list(dict.fromkeys(stem_nltk))
len(stem_nltk)
```

### Lemmatization

NLTK has lemmatization.

Spacy has lemmatization.

Textblob has lemmatization.

```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4') #download omw-1.4 as a prereq for wordnet
nltk.download('wordnet') #download Wordnet data for the lemmatization
nltk_lemmatizer = WordNetLemmatizer()
list_of_words = ['run','runner','running','ran','runs','easily','fairly', "cook","cooker","cooking","cooked"]
nltk_example = [nltk_lemmatizer.lemmatize(word) for word in list_of_words]
nltk_example
```

```python
#here is the stemming example again, what is the difference?
stem_example
```

```python
lemmatize_nltk = [nltk_lemmatizer.lemmatize(x) for x in tokenized_text_nltk_wo_sw] #use list comprehension to stem
lemmatize_nltk
```

```python
lemmatize_nltk = list(dict.fromkeys(lemmatize_nltk))  #remove duplicates
len(lemmatize_nltk)
```
